```python
import zipfile
import os
import subprocess
import pandas as pd

data_path = 'Z:\DEV\weather-forcast\CCTV_DATA'
train_label_data_path = data_path+'/01-1.정식개방데이터/Training/02.라벨링데이터/'
vak_label_data_path = data_path+'/01-1.정식개방데이터/Validation/02.라벨링데이터/'

# 파일 리스트 만들기
## total_file_list 를 만듦
def find_files_in_folder(folder_path):
    file_list = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            file_list.append(os.path.join(root, file))
    return file_list
result = subprocess.check_output("dir "+data_path+" /B /AD", shell=True)
directories = result.decode('utf-8', errors='ignore').splitlines()
folder_list = [directory for directory in directories if directory.strip()]
print(folder_list)
total_file_list = []
for folder in folder_list:
    folder_path = os.path.join(data_path, folder)
    file_list = find_files_in_folder(folder_path)
    for i in file_list:
        print(i)
        total_file_list.append(i)


# 압축 풀기
## total_file_list를 반복
for zip_file in total_file_list:
    extract_dir = os.path.dirname(zip_file)
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)


# json 파일 목록을 만듦
## json_list : json 파일 이름 리스트
result = subprocess.check_output("dir "+data_path+"/01-1.정식개방데이터/Training/02.라벨링데이터 /B /A-D", shell=True)
directories = result.decode('utf-8', errors='ignore').splitlines()
json_list = []
for i in directories:
    if i[-5:] == '.json':
        json_list.append(i)


# json을 읽어서 필요한 데이터만 추출
df = pd.read_json(train_data_path+json_list[0])
print(train_data_path+json_list[0])
data = {'filename':[df['Source_Data_Info'].source_data_ID],
        'path':[train_data_path+json_list[0]],
        'weather':[df['Raw_Data_Info'].weather],
        'resolution':[df['Raw_Data_Info'].resolution],
        'time':[df['Source_Data_Info'].extract_time],
        'date':[df['Raw_Data_Info'].date],
        'day_type':[df['Raw_Data_Info'].day_type],
        'season':[df['Raw_Data_Info'].season]}
set_df = pd.DataFrame(data=data)
set_df
```
